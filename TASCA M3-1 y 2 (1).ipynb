```python
#TASCA M3-1.3
#Crea una función que dada una tabla de dos dimensiones (NxM), te calcule los totales por fila y los totales por columna.

#PRIMERA FORMA

matriz = [
    [1, 2, 6],
    [6, 9, 6],
    [7, 8, 9]
]

suma_filas = []

for f in matriz:
    suma_filas.append(sum(f))

print(suma_filas) 

suma_columnas = []

for i in matriz:
    suma_columnas.append(sum(i))

print(suma_columnas)

#SEGUNDA FORMA

def mostrar_matriz(matriz):
    for columna in matriz:
        print(columna)


matriz = [
    [25, 69, 51, 26],
    [68, 35, 29, 54],
    [54, 57, 45, 63],
    [61, 68, 47, 59]
]

mostrar_matriz(matriz)

filas = len(matriz)
columnas = len(matriz)

for f in range(columnas):
    suma = sum(matriz[f])
    matriz[f].append(suma)

print()

mostrar_matriz(matriz)


```
